## Exploration of Simpson's Rule

Or this tiny application of it, anyway

```python
from __future__ import print_function
from sympy import nonlinsolve, linsolve, symbols, latex, expr

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def f(x):
    return x**2+3*x+5

p0 = (-4, f(-4))
p1 = (0, f(0))
p2 = (4, f(4))

areaExpr = ((p2[0]-p0[0])/6)*(p0[1]+4*p1[1]+p2[1])

a, b, c, x = symbols('a b c x')

def quadSub(x, y):
    return a*x**2+b*x+c-y

soe = [
    quadSub(p0[0], p0[1]),
    quadSub(p1[0], p1[1]),
    quadSub(p2[0], p2[1]),
]

solved = nonlinsolve(soe, [a, b, c])

solvedCoefficients = tuple(*solved)

import matplotlib.pyplot as plt
import numpy as np

padding = 2;

min_x = p0[0] - padding;
max_x = p2[0] + padding;

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(min_x,max_x,100)

a, b, c = solvedCoefficients

expr = a*x**2+b*x+c

def eval_expr(x):
    return float(a)*x**2+float(b)*x+float(c)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.plot(x, expr, ':b', label=f'${a}x^2 + {b}x + {c}$')
plt.plot(p0[0], p0[1], 'o')
plt.plot(p1[0], p1[1], 'o')
plt.plot(p2[0], p2[1], 'o')
section = np.arange(p0[0], p2[0], 1/50.)
plt.fill_between(section, 0, eval_expr(section), facecolor='blue', alpha=0.5)

plt.legend(loc='upper left')
plt.show()

(Fraction(areaExpr).limit_denominator(), p0[1], 4*p1[1], p2[1], p0[1]+4*p1[1]+ p2[1], Fraction((p2[0]-p0[0])/6).limit_denominator())
```
